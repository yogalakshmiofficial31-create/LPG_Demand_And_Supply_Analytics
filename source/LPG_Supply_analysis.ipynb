#**Demand by State**

```python
state_demand = LPG.groupby("State")["Cylinders_Requested"].sum()

state_demand
```

```python
state_demand.sort_values().plot(kind="bar", figsize=(10,5))

plt.title("LPG Demand by State")
plt.xlabel("State")
plt.ylabel("Cylinders Requested")

plt.xticks(rotation=45)

plt.show()
```

#**Monthly Demand**

```python
monthly = LPG.groupby("Month")["Cylinders_Requested"].sum()

monthly
```

```python
monthly.plot(kind="line", marker="o", figsize=(10,5))

plt.title("Monthly LPG Demand")
plt.xlabel("Month")
plt.ylabel("Cylinders Requested")

plt.show()
```

#**Festival Impact**

```python
festival = LPG.groupby("Festival")["Cylinders_Requested"].mean()

festival
```

```python
festival.plot(kind="bar", figsize=(8,5))

plt.title("Festival Impact on LPG Demand")

plt.show()
```

#**Customer Type Analysis**

```python
customer = LPG.groupby("Customer_Type")[["Revenue","Profit"]].sum()

customer
```

```python
customer.plot(kind="bar", figsize=(8,5))

plt.title("Revenue & Profit by Customer Type")

plt.show()
```

#**Iran–Israel Conflict Analysis**

```python
conflict = LPG.groupby("Iran_Israel_Conflict_Level")[["Oil_Price_USD","Shipment_Delay_Days"]].mean()

conflict
```

```python
conflict.plot(kind="bar", figsize=(8,5))

plt.title("Conflict Level vs Oil Price & Shipment Delay")

plt.show()
```

#Shipment Delay vs Complaints

```python
plt.figure(figsize=(8,5))

plt.scatter(LPG["Shipment_Delay_Days"], LPG["Consumer_Complaints"])

plt.title("Shipment Delay vs Consumer Complaints")

plt.xlabel("Shipment Delay (Days)")
plt.ylabel("Consumer Complaints")

plt.show()
```
