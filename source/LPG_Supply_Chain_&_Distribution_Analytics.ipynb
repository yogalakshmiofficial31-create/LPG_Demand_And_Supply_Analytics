#**1.Which states have the highest demand for LPG cylinders, and how does this help us prioritize our delivery resources?**

```python
# Aggregate and sort demand in descending order for better top-to-bottom reading
state_demand = df.groupby("State")["Cylinders_Requested"].sum().sort_values(ascending=True)

# Initialize chart layout
plt.figure(figsize=(10, 6.5))

# Generate a professional color gradient (darker colors for higher demand)
colors = sns.color_palette("YlOrRd", len(state_demand))

bars = plt.barh(
    state_demand.index,
    state_demand.values,
    color=colors,
    edgecolor="#333333",
    linewidth=0.7
)

# Overlay precise volume metrics on bar ends with clean formatting
for bar in bars:
    width = bar.get_width()
    plt.text(
        width + (state_demand.max() * 0.01),  # Dynamic text padding
        bar.get_y() + bar.get_height()/2,
        f" {width:,.0f}",
        va="center",
        ha="left",
        fontsize=9,
        fontweight="semibold",
        color="#2c3e50"
    )

# Professional chart properties
plt.title("Total LPG Cylinder Demand by State", fontsize=14, fontweight="bold", pad=15)
plt.xlabel("Total Cylinders Requested", fontsize=11, labelpad=10)
plt.ylabel("State Market Profile", fontsize=11)

# Style grids and spine borders
plt.grid(axis="x", linestyle="--", alpha=0.4)
sns.despine(left=True, bottom=True)

# Dynamically expand X-axis limits to prevent text clipping
plt.xlim(0, state_demand.max() * 1.12)

plt.tight_layout()
plt.show()
```

Top Market Focus: The chart clearly shows which states order the most cylinders. must prioritize these high-volume states to prevent stock shortages.
Resource Planning: We can use these demand numbers to decide how many delivery trucks and cylinders to assign to each state.

#**2.Which customer segment drives the highest product demand, and how should this influence our distribution priority?**

```python
# Aggregate and sort customer demand
customer_demand = df.groupby("Customer_Type")["Cylinders_Requested"].sum().sort_values(ascending=False)

plt.figure(figsize=(7, 5))

# clean, professional palette (blue for top, grey for secondary)
colors = ['#1f77b4', '#aec7e8'] if len(customer_demand) >= 2 else ['#1f77b4']

bars = plt.bar(
    customer_demand.index,
    customer_demand.values,
    color=colors,
    edgecolor="#333333",
    linewidth=0.8,
    width=0.5  # Cleaner bar width
)

# clear formatting values on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,
        height + (customer_demand.max() * 0.01),  # Dynamic vertical padding
        f"{height:,.0f}",
        ha="center",
        va="bottom",
        fontsize=10,
        fontweight="semibold",
        color="#2c3e50"
    )

# Clean and bold title structure
plt.title("LPG Cylinder Demand by Customer Type", fontsize=13, fontweight="bold", pad=15)
plt.xlabel("Customer Segment", fontsize=11, labelpad=8)
plt.ylabel("Total Cylinders Requested", fontsize=11)

# Remove distracting box lines (spines)
plt.grid(axis="y", linestyle="--", alpha=0.4)
sns.despine(left=True, bottom=False)

# top padding to the chart
plt.ylim(0, customer_demand.max() * 1.12)

plt.tight_layout()
plt.show()
```

#**3.How do delivery delays impact customer complaints, and what is the exact breaking point for our service quality?**

```python
plt.figure(figsize=(8, 5.5))

# seaborn regplot to combine scatter and trend line automatically
sns.regplot(
    data=df,
    x="Shipment_Delay_Days",
    y="Consumer_Complaints",
    scatter_kws={"alpha": 0.6, "edgecolors": "black", "linewidths": 0.5, "s": 50, "color": "#1f77b4"},
    line_kws={"color": "#d62728", "linewidth": 2.5}  # Sharp red for the trend line
)

# Calculate Pearson correlation coefficient dynamically
correlation = df["Shipment_Delay_Days"].corr(df["Consumer_Complaints"])

# correlation text onto the chart area
plt.text(
    df["Shipment_Delay_Days"].min() + 1,
    df["Consumer_Complaints"].max() * 0.9,
    f"Correlation: {correlation:.2f}",
    fontsize=11, fontweight="bold", color="#d62728",
    bbox=dict(facecolor='white', alpha=0.8, edgecolor='none')
)

# Professional titles and layout cleanups
plt.title("Delivery Delays vs Consumer Complaints", fontsize=13, fontweight="bold", pad=15)
plt.xlabel("Shipment Delay (Days)", fontsize=11, labelpad=8)
plt.ylabel("Number of Consumer Complaints", fontsize=11)

plt.grid(linestyle="--", alpha=0.4)
sns.despine()

plt.tight_layout()
plt.show()
```

Direct Cause and Effect: It means that 98% of the movement in customer complaints can be explained directly by delivery delays.

Predictable Pattern: If delivery delays go up tomorrow, you can predict almost exactly how many customer complaints your support team will receive.

#**4.How do global conflicts and shipping route disruptions impact our domestic delivery timelines?**

```python
# Create a risk impact matrix
geopolitical_impact = df.groupby(['Iran_Israel_Conflict_Level', 'Strait_of_Hormuz_Status']).agg({
    'Oil_Price_USD': 'mean',
    'Shipment_Delay_Days': 'mean'
}).reset_index()

pivot_delays = geopolitical_impact.pivot(index="Iran_Israel_Conflict_Level",
                                         columns="Strait_of_Hormuz_Status",
                                         values="Shipment_Delay_Days")

plt.figure(figsize=(8, 5))
sns.heatmap(pivot_delays, annot=True, cmap='YlOrRd', fmt=".1f", cbar_kws={'label': 'Avg Delay (Days)'})
plt.title('Average Delivery Delays vs Geopolitical Chokepoint Risk')
plt.show()
```

#**5.How do changes in global oil prices influence domestic LPG prices, government subsidies, and our bottom-line profits?**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# 1. Compute macro correlation matrix
macro_cols = ['Oil_Price_USD', 'Domestic_LPG_Price', 'Subsidy_Amount', 'Profit']
macro_corr = df[macro_cols].corr()

plt.figure(figsize=(8, 6))

# 2. Plot refined correlation heatmap
sns.heatmap(
    macro_corr,
    annot=True,
    cmap='coolwarm',
    fmt=".2f",
    vmin=-1,
    vmax=1,
    linewidths=1,
    linecolor='white',
    annot_kws={"size": 11, "weight": "bold"},
    square=True  # Forces the cells to be perfect squares
)

# Professional titles and axis label adjustments
plt.title('Financial Linkages: Global Oil Index vs. Domestic LPG Markets', fontsize=13, fontweight='bold', pad=15)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(rotation=0, fontsize=10)

plt.tight_layout()
plt.show()

```
